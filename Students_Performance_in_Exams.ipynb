```python
def pie_graph(df, value, title):
  labels = df[value].value_counts().index
  values = df[value].value_counts()

  fig = go.Figure(data = [go.Pie(labels=labels, values = values, hole= .5)])
  fig.update_layout(title_text=title)
  fig.show()
```

```python
pie_graph(df, 'gender', 'Gender')
```

```python
pie_graph(df, 'race/ethnicity', 'race/ethnicity')
```

```python
pie_graph(df, 'parental level of education', 'Parental level of education')
```

```python
pie_graph(df, 'lunch', 'Lunch')
```

```python
pie_graph(df, 'test preparation course', 'Test preparation course')
```
